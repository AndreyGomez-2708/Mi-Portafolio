```
# 1. IMPORTAR LAS LIBRERÍAS QUE VOY A NECESITAR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Para que no me salgan warnings fastidiosos
warnings.filterwarnings('ignore')

# Configuración para que las gráficas se vean bien
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

print("Librerías importadas correctamente\n")
```

```
# 2. CARGAR EL DATASET

print("CARGANDO EL DATASET")

# enlace del dataset de UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# nombres de las columnas
columnas = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs',
    'restecg', 'thalach', 'exang', 'oldpeak',
    'slope', 'ca', 'thal', 'target'
]

# cargo el dataset y le digo que los '?' son valores nulos
df = pd.read_csv(url, names=columnas, na_values='?')

print(f"\n✓ Dataset cargado exitosamente")
print(f"  Filas: {df.shape[0]}")
print(f"  Columnas: {df.shape[1]}")

# miro cómo se ven los primeros datos
print("\nPrimeras 5 filas del dataset:")
print(df.head())

print("\nInformación general:")
print(df.info())

print("\nEstadísticas básicas:")
print(df.describe())
```

```
# 3. LIMPIEZA DE DATOS

print("LIMPIEZA DE DATOS")

# Primero veo si hay valores nulos
print("\n1. Revisando valores nulos:")
nulos = df.isnull().sum()
print(nulos[nulos > 0])

filas_con_nulos = df.isnull().any(axis=1).sum()
porcentaje_nulos = (filas_con_nulos / len(df)) * 100

print(f"\nTotal de filas con nulos: {filas_con_nulos}")
print(f"Esto representa el {porcentaje_nulos:.2f}% de los datos")

print("""
MI DECISIÓN SOBRE LOS NULOS:
Veo que las columnas 'ca' y 'thal' tienen algunos valores nulos, pero son
muy pocos (como 6 filas). Como estas variables son importantes para detectar
enfermedades del corazón, decidí eliminar esas filas en vez de rellenarlas
con promedios o algo así, porque podría meter ruido en el análisis.
Además, son tan pocas filas que no afecta mucho el análisis general.
""")

# Elimino las filas con nulos
df_limpio = df.dropna()
print(f"Filas después de limpiar: {len(df_limpio)}")

# Ahora reviso si hay valores raros (outliers)
print("\n2. Buscando valores atípicos:")

columnas_numericas = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

for col in columnas_numericas:
    Q1 = df_limpio[col].quantile(0.25)
    Q3 = df_limpio[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_bajo = Q1 - 1.5 * IQR
    limite_alto = Q3 + 1.5 * IQR

    outliers_bajos = (df_limpio[col] < limite_bajo).sum()
    outliers_altos = (df_limpio[col] > limite_alto).sum()

    if outliers_bajos + outliers_altos > 0:
        print(f"  {col}: {outliers_bajos} valores muy bajos, {outliers_altos} valores muy altos")

print("""
MI DECISIÓN SOBRE LOS OUTLIERS:
Encontré algunos valores extremos en colesterol y presión arterial, pero
decidí NO eliminarlos porque en temas de salud, esos casos extremos son
justamente los más interesantes de analizar. Un colesterol altísimo o una
presión muy alta son precisamente los que tienen más riesgo de enfermedad.
""")

# Arreglo la variable target para que sea más fácil de trabajar
print("\n3. Ajustando la variable objetivo (target):")
print(f"Valores originales: {sorted(df_limpio['target'].unique())}")

print("""
La variable 'target' tiene valores de 0 a 4, donde:
- 0 = sin enfermedad
- 1, 2, 3, 4 = diferentes niveles de enfermedad

Para simplificar el análisis, voy a convertirlo a binario:
- 0 = NO tiene enfermedad
- 1 = SÍ tiene enfermedad (cualquier nivel)
""")

df_limpio['target'] = (df_limpio['target'] > 0).astype(int)
print(f"\n Variable convertida a binaria")
print(f"Distribución final:")
print(df_limpio['target'].value_counts())
```

```
# 4. FILTROS Y SUBGRUPOS

print("CREANDO FILTROS PARA ANALIZAR SUBGRUPOS")

# Filtro 1: Mayores de 50 años
mayores_50 = df_limpio[df_limpio['age'] > 50].copy()
print(f"\n1. Pacientes mayores de 50 años: {len(mayores_50)}")
print(f"   Con enfermedad: {mayores_50['target'].sum()} ({mayores_50['target'].mean()*100:.1f}%)")

# Filtro 2: Mujeres con hipertensión (sex=0 es mujer, PA>140 es hipertensión)
mujeres_hipertension = df_limpio[(df_limpio['sex'] == 0) & (df_limpio['trestbps'] > 140)].copy()
print(f"\n2. Mujeres con hipertensión (presión > 140): {len(mujeres_hipertension)}")
if len(mujeres_hipertension) > 0:
    print(f"   Con enfermedad: {mujeres_hipertension['target'].sum()} ({mujeres_hipertension['target'].mean()*100:.1f}%)")

# Filtro 3: Colesterol alto (>240 es alto)
colesterol_alto = df_limpio[df_limpio['chol'] > 240].copy()
print(f"\n3. Pacientes con colesterol alto (> 240): {len(colesterol_alto)}")
print(f"   Con enfermedad: {colesterol_alto['target'].sum()} ({colesterol_alto['target'].mean()*100:.1f}%)")

# Filtro 4: Grupo de alto riesgo (combinación de factores)
alto_riesgo = df_limpio[
    (df_limpio['age'] > 55) &
    (df_limpio['chol'] > 240) &
    (df_limpio['trestbps'] > 140)
].copy()
print(f"\n4. Pacientes de ALTO RIESGO (>55 años + colesterol alto + hipertensión):")
print(f"   Total: {len(alto_riesgo)}")
if len(alto_riesgo) > 0:
    print(f"   Con enfermedad: {alto_riesgo['target'].sum()} ({alto_riesgo['target'].mean()*100:.1f}%)")
```

```
# 5. TRANSFORMACIONES Y FUNCIONES

print("APLICANDO TRANSFORMACIONES")

# Creo categorías para edad
def categoria_edad(edad):
    """Pongo las edades en grupos para analizar mejor"""
    if edad < 40:
        return '< 40'
    elif edad < 50:
        return '40-49'
    elif edad < 60:
        return '50-59'
    else:
        return '60+'

df_limpio['grupo_edad'] = df_limpio['age'].apply(categoria_edad)

# Categorizo el colesterol
def categoria_colesterol(chol):
    """Según estándares médicos básicos"""
    if chol < 200:
        return 'Normal'
    elif chol < 240:
        return 'Limítrofe'
    else:
        return 'Alto'

df_limpio['nivel_colesterol'] = df_limpio['chol'].apply(categoria_colesterol)

# Categorizo la presión arterial
def categoria_presion(pa):
    """Niveles de presión arterial"""
    if pa < 120:
        return 'Normal'
    elif pa < 140:
        return 'Elevada'
    else:
        return 'Hipertensión'

df_limpio['nivel_presion'] = df_limpio['trestbps'].apply(categoria_presion)

print("\n✓ Categorías creadas exitosamente")

# Ahora saco estadísticas por grupos
print("\n1. Estadísticas por SEXO (0=Mujer, 1=Hombre):")
stats_sexo = df_limpio.groupby('sex').agg({
    'age': ['mean', 'std'],
    'chol': ['mean', 'std'],
    'trestbps': ['mean', 'std'],
    'target': ['sum', 'mean']
}).round(2)
print(stats_sexo)

print("\n2. Estadísticas por GRUPO DE EDAD:")
stats_edad = df_limpio.groupby('grupo_edad').agg({
    'chol': 'mean',
    'trestbps': 'mean',
    'target': ['sum', 'mean', 'count']
}).round(2)
print(stats_edad)

print("\n3. Estadísticas por NIVEL DE COLESTEROL:")
stats_col = df_limpio.groupby('nivel_colesterol').agg({
    'age': 'mean',
    'target': ['sum', 'mean', 'count']
}).round(2)
print(stats_col)
```

```
# 6. TABLA CONSOLIDADA

print("TABLA CONSOLIDADA FINAL")

# Creo una tabla resumen con lo más importante
tabla_final = df_limpio.groupby(['grupo_edad', 'sex']).agg({
    'age': ['count', 'mean'],
    'chol': 'mean',
    'trestbps': 'mean',
    'thalach': 'mean',
    'target': ['sum', 'mean']
}).round(2)

tabla_final.columns = [
    'total', 'edad_promedio', 'colesterol_promedio',
    'presion_promedio', 'fc_maxima_promedio',
    'total_enfermos', 'tasa_enfermedad'
]

tabla_final = tabla_final.reset_index()
tabla_final['sex'] = tabla_final['sex'].map({0: 'Mujer', 1: 'Hombre'})

print("\nResumen por Edad y Sexo:")
print(tabla_final)

# Resumen general
print("\n" + "="*70)
print("RESUMEN GENERAL")
print("="*70)
total = len(df_limpio)
enfermos = df_limpio['target'].sum()
sanos = total - enfermos
print(f"""
Total de pacientes: {total}
Con enfermedad: {enfermos} ({enfermos/total*100:.1f}%)
Sin enfermedad: {sanos} ({sanos/total*100:.1f}%)

Hombres: {(df_limpio['sex']==1).sum()} ({(df_limpio['sex']==1).sum()/total*100:.1f}%)
Mujeres: {(df_limpio['sex']==0).sum()} ({(df_limpio['sex']==0).sum()/total*100:.1f}%)

Edad promedio: {df_limpio['age'].mean():.1f} años
Colesterol promedio: {df_limpio['chol'].mean():.1f} mg/dl
Presión promedio: {df_limpio['trestbps'].mean():.1f} mmHg
""")
```

```
# EJERCICIO BONIFICACIÓN

print("EJERCICIO BONIFICACIÓN - Calculadora de Riesgo")


def calcular_riesgo(edad, sexo, presion, colesterol, fuma=False):
    """
    Función que calcula el riesgo cardiovascular de una persona.

    Parámetros:
    - edad: años
    - sexo: 0 para mujer, 1 para hombre
    - presion: presión arterial en mmHg
    - colesterol: en mg/dl
    - fuma: True o False

    Devuelve un diccionario con el nivel de riesgo y recomendaciones
    """
    puntos = 0

    # Sumo puntos por edad (más viejo = más puntos)
    if sexo == 1:  # Hombre
        if edad >= 70:
            puntos += 5
        elif edad >= 60:
            puntos += 4
        elif edad >= 50:
            puntos += 3
        elif edad >= 40:
            puntos += 2
    else:  # Mujer
        if edad >= 70:
            puntos += 6
        elif edad >= 60:
            puntos += 5
        elif edad >= 50:
            puntos += 3
        elif edad >= 40:
            puntos += 1

    # Puntos por presión arterial
    if presion >= 160:
        puntos += 3
    elif presion >= 140:
        puntos += 2
    elif presion >= 120:
        puntos += 1

    # Puntos por colesterol
    if colesterol >= 280:
        puntos += 3
    elif colesterol >= 240:
        puntos += 2
    elif colesterol >= 200:
        puntos += 1

    # Puntos extra si fuma
    if fuma:
        puntos += 2

    # Determino el nivel de riesgo según los puntos
    if puntos <= 4:
        nivel = "BAJO"
        porcentaje = "menos del 10%"
        recomendacion = "Todo bien. Mantén hábitos saludables y hazte chequeos cada 2 años."
    elif puntos <= 8:
        nivel = "MODERADO"
        porcentaje = "entre 10-20%"
        recomendacion = "Ojo, hay que cuidarse más. Mejora tu alimentación y haz ejercicio. Chequeo anual."
    elif puntos <= 12:
        nivel = "ALTO"
        porcentaje = "entre 20-30%"
        recomendacion = "ALERTA. Necesitas ir al cardiólogo. Probablemente necesites medicación."
    else:
        nivel = "MUY ALTO"
        porcentaje = "más del 30%"
        recomendacion = "URGENTE. Debes ir al médico YA. Riesgo muy alto de infarto."

    return {
        'puntos': puntos,
        'nivel': nivel,
        'riesgo': porcentaje,
        'que_hacer': recomendacion
    }


# Pruebo la función con algunos pacientes del dataset
print("\nProbando la calculadora con pacientes reales del dataset:\n")

if len(df_limpio) >= 5:
    muestra = df_limpio.sample(5, random_state=42)
else:
    muestra = df_limpio.head(5)

for idx, paciente in muestra.iterrows():
    resultado = calcular_riesgo(
        edad=int(paciente['age']),
        sexo=int(paciente['sex']),
        presion=float(paciente['trestbps']),
        colesterol=float(paciente['chol']),
        fuma=False  # no tenemos este dato en el dataset
    )

    print(f"Paciente #{idx}:")
    print(f"  {int(paciente['age'])} años, {'Hombre' if paciente['sex']==1 else 'Mujer'}")
    print(f"  Presión: {paciente['trestbps']:.0f} | Colesterol: {paciente['chol']:.0f}")
    print(f"  Enfermedad real: {'SÍ' if paciente['target']==1 else 'NO'}")
    print(f"  ➜ RIESGO {resultado['nivel']} ({resultado['riesgo']})")
    print(f"  ➜ {resultado['que_hacer']}")
    print()

print("Ejercicio de bonificación completado con buenas prácticas PEP8 ;D")
```
